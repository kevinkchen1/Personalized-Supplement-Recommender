```python
import pandas as pd
side_effects = pd.read_csv('side_effects.csv')
supplements = pd.read_csv('supplements.csv')
symptoms = pd.read_csv('symptoms_list.csv')

res = pd.merge(side_effects, supplements, on = 'supplement_name')
res = pd.merge(res, symptoms, on = 'symptom_name')

res = res[['supplement_id', 'symptom_id']]
res.head()


res.to_csv('side_effects_supplements_symptoms.csv', index = False)

```
